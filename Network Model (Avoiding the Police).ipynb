## NetworkX Implementation

```python
#import the networkx library
import networkx as nx
```

```python
#setting up the network system
G=nx.Graph()
weight=[('1','2',0.69897),('1','3',0.04576), #add the weights(probabilities)
        ('2','3',0.22185),('2','4',0.09691),
        ('3','4',1.0),('3','5',0.52288),
        ('4','5',0.39794),('4','6',0.45593),
        ('5','7',0.60206),('6','7',0.30103)]

G.add_weighted_edges_from(weight)

# Print all possible paths to end node
list(nx.shortest_simple_paths(G,'1','7'))
```

```python
#Find the shortest route
print('Shortest route from 1 to 7 is {}'.format(list(nx.shortest_path(G,'1','7'))))
```

The optimal route that maximizes the probability of not being fined is through nodes 1, 3, 5, and 7 which will yield a probability of **0.0675**.
